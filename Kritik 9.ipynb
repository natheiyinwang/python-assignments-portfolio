```python
import sympy as sp 
```

```python
#defining symbols
x, y = sp.symbols('x y') 

#defining function
f = sp.exp(x)*sp.sin(y) +y**3 

#diff
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y) 

print (df_dx,",", df_dy) 


```

```python
#gradient vector
x, y = sp.symbols('x y') 
g = (x**2)*y + x*y**2


#diff
dg_dx = sp.diff(g,x) 
dg_dy = sp.diff(g,y) 

x_sub = dg_dx.subs({x:1, y:-1})
y_sub = dg_dy.subs({x:1, y:-1}) 

g_sub = sp.sqrt((x_sub)**2 + (y_sub)**2)



print("The gradient vector is", "(", x_sub, ",", y_sub, ")" ,",", "and its magnitude is", g_sub)


```

```python
#second derivative

x,y = sp.symbols ('x y') 

h = sp.log((x**2) + (y**2)) 

dh_dxx = sp.diff(h,x,x) 
dh_dyy = sp.diff(h,y,y) 
dh_dxy = sp.diff(h,y,x)

print( dh_dxx, dh_dyy, dh_dxy, "According to Clairaut's theorem, if the mixed second order derivatives are continuous in the region (a,b), then the derivatives are symmetric.")

```

```python
#contour plot
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
#my own function 
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = (1/(15*(x**2 + y**2)))

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-0.5, 0.5, 50)
y_vals = np.linspace(-0.5, 0.5, 50)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = 1/(15(x^2+y^2))$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
